## Matplotlib example and get it ready for potting in JP

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 4]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #creates hight and width 

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot", # title of the pic 
       xlabel="x-axis",    # x=width
       ylabel="y-axis")    # y=hight

# 5. Save and show
fig.savefig("sample_plot.png")
plt.show()
```

# Make a plot from dict

```python
car_price = {"M.Benz" : 10.000,
            "BMW": 20.000,
            "Lanborghini": 90.000}
fig, ax = plt.subplots()
ax.bar(car_price.keys(), car_price.values())
ax.set(title = "Car Prices in 2024",
      ylabel= "Price in $",
      xlabel = "Car brand")
```

```python
fig, ax = plt.subplots()
ax.barh(list(car_price.keys()), list(car_price.values()));
ax.set(title = "Car Prices in 2024",
      ylabel= "Price in $",
      xlabel = "Car brand")
```

# Two options for subplots

```python
# First option
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows =2,
                                          ncols =2,
                                          figsize=(10,5))
ax1.plot(x, x/2);
ax2.scatter(np.random.rand(10), np.random.rand(10));
ax3.bar(car_price.keys(), car_price.values());
ax4.hist(np.random.rand(100));
```

```python
# Second option

fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))

# Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.rand(10), np.random.rand(10));
ax[1,0].bar(car_price.keys(), car_price.values());
ax[1,1].hist(np.random.rand(100));
```

# Another file 

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

# PYPLOT method

```python
over_50.plot(kind ="scatter",
            x="age",
            y="chol",
            c="target",
             figsize = (10,6)
            )
```

# OO method

```python
fig, ax = plt.subplots(figsize=(10,7))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax = ax);
ax.set_xlim([45,100]);
```

```python
fig, ax = plt.subplots(figsize=(10,6))

# plot the data 
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

ax.set(title = "Heart Desease Data",
      xlabel="Age",
      ylabel="Cholestrerol")
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")


# Add a line 
ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

# Subplot of chol, age, thalach

```python
fig, (ax0, ax2) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                           sharex=True
                              )


# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
ax0.set(title = "Heart disease and Cholesterol levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a legend 
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle = "--");

# do it on AX2

scatter2 = ax2.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
ax2.set(title="Thalach level",
       xlabel="Age",
       ylabel="Thalach");

# Add a legend 
ax2.legend(*scatter.legend_elements(), title="Target");

# Add a line
ax2.axhline(y=over_50["thalach"].mean(),
           linestyle = "--");

# Add a title to the fig 
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")
```
