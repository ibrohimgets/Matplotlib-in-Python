# 1st method

```python
fig = plt.figure() # creates a figure
ax = fig.add_subplot() #adds some axes
plt.show()
```

# 2nd method

```python
fig = plt.figure() # creates a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x,y)
plt.show()
```

# 3rd method

```python
fig, ax = plt.subplots()
ax.plot(x,y);
```

## Matplotlib example and get it ready for potting in JP

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 4]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #creates hight and width 

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot", # title of the pic 
       xlabel="x-axis",    # x=width
       ylabel="y-axis")    # y=hight

# 5. Save and show
fig.savefig("sample_plot.png")
plt.show()
```
