```python
over_50 = pd.read_csv("heart-disease.csv")
```

```python
#style 
plt.style.use("classic")


fig, ax = plt.subplots(figsize=(10,6))

# plot the data 
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="summer");

ax.set(title = "Heart Desease Data",
      xlabel="Age",
      ylabel="Cholestrerol")
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")


# Add a line 
ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

# style the x and y

```python
fig, (ax0, ax2) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                           sharex=True
                              )


# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap=("winter"));

ax0.set(title = "Heart disease and Cholesterol levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a legend 
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle = "--");


# do it on AX2

scatter2 = ax2.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                       cmap=("winter"))
ax2.set(title="Thalach level",
       xlabel="Age",
       ylabel="Thalach");

# Add a legend 
ax2.legend(*scatter.legend_elements(), title="Target");

# Add a line
ax2.axhline(y=over_50["thalach"].mean(),
           linestyle = "--");


# Add a title to the fig 
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")

# Save this as an Image
plt.savefig("heart_disease_analysis.png", dpi=300, bbox_inches='tight')
```
